```python
#quest 10b
#Inheritance

#parent Class
class Adubuola:
    def __init__(self):
        print("An Adubuola Was Born")
        
    def whoisThis(self):
        print("An Adubuola")
        
    def gender(self):
        print("Male/Female")
        
#child class
class Femi(Adubuola):
    def __init__(self):
        #taking the parent super function
        super().__init__()
        print("Named Adubuola Femi")
    
    def whoisThis(self):
        print("Femi")
    
    def gender(self):
        print("Male")
    def move(self):
        print("Moving...")
        
anonymous = Femi()
anonymous.whoisThis()
anonymous.gender()
anonymous.move()
```

```python
#quest 10b
#Inheritance

#parent Class
class Animal:
    def __init__(self):
        print("Animal is Ready")
        
    def whoisThis(self):
        print("Animal")
        
    def eat(self):
        print("Eating...")
        
    def jump(self):
        print("Jumping...")
        
    def run(self):
        print("Run Faster")
        
#child class
class Bird(Animal):
    def __init__(self):
        #taking the parent super function
        super().__init__()
        print("Bird is Ready")
    
    def whoisThis(self):
        print("Bird")
    
    
        
hawk = Bird()
hawk.whoisThis()
hawk.eat()
hawk.jump()
hawk.run()
```
